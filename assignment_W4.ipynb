## 1. Read the dataset to the python environment.

```python
import os
import numpy as np
import pandas as pd

pwd=os.getcwd()
cars=pd.read_csv(pwd+"/cars_data.csv")
cars
```

## 2. Check for the null values present in the dataset.

```python
cars.isnull().sum()
```

## 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
import matplotlib.pyplot as plt
import seaborn as sns

graph = cars["Buyer Gender"].value_counts()
graph.plot(kind='bar')

plt.title('No.of Male and Female Buyers')
plt.ylabel('Total Numbers') 
plt.show()

```

## 4. Find the top 5 cars based on their sales price.

```python
top5=cars.sort_values(by=['Sale Price']).head(5)
top5['Make']+" "+top5['Model']
```

## 5. Find the least 5 cars based on their Resell price

```python
last5=cars.sort_values(by=['Resell Price']).tail(5)
last5['Make']+" "+last5['Model']
```
