## Q5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

### BAR GRAPH

```python
import matplotlib.pyplot as plt
import seaborn as sns

type = df.groupby('Classification')['SL'].mean().to_frame().reset_index()

plt.bar(type['Classification'],type['SL'],color = ['red','green','skyblue']) 

plt.title('Bar Chart')
plt.xlabel('Classification')
plt.ylabel('SepalLength') 
plt.show()
```

### LINE PLOT


```python
df_values =df.iloc[:,:-1]
columns = df_values.columns
# create x data
x_value = range(0, 150)
# create figure and axis
fig,ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_value, df[column], label=column)
ax.legend()
```

### SCATTER PLOTS


```python
plt.scatter(df['SL'],df['PL'],alpha=1) 
plt.title('Scatter Plot')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')

plt.show()
```
