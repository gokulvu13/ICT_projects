## <font color=red> 2. Build a regression model to predict the profit from the given data and comment your findings on it.

```python
sales1=sales.drop('Date',axis=1) # dropping unnecessary columns
sales1.columns
```

```python
y = sales['Profit(in Thousands)'] # the target 
X = sales1.drop(['Profit(in Thousands)'], axis=1) #decided to include all the features 
X.shape
```

```python
# Seperating the data into test and train 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.25) 
```

```python
from sklearn import linear_model
lr= linear_model.LinearRegression()
model = lr.fit(X_train, y_train)

predictions = model.predict(X_test)
print(y_test)
```

```python
print(predictions)

from sklearn.metrics import mean_squared_error

print('MSE is:',mean_squared_error(y_test,predictions))
```

#### The MSE is minimal and very close to zero
