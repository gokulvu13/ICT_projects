```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp

df=pd.read_csv('Sales_data_zone_wise.csv')
df.head()
```

<font color=magenta>Objective 1: To find the mean sales generated by each zone.

```python
df.mean().round(2)
```

<font color=magenta>Objective 2: To find total sales generated by all the zones for each month.

```python
zone=list(df)
zone.remove("Month")
df["Total Sale"]=df[zone].sum(axis=1)
df.head()
```

<font color=magenta>Objective 3 : Check whether all the zones generate the same amount of sales

Using One way Anova Test because there are three groups 

Null Hypothesis(H0): the zones generate the same amount of sales

Alternate Hypothesis(H1): the zones  generate different amount of sales.

```python
from scipy import stats

f, p = stats.f_oneway(df['Zone - A'], df['Zone - B'], df['Zone - C'],df['Zone - D'])
print('F test statistic:',f,'p-value:',p)
```

```python
if p >= 0.05:
    print('Null Hypothesis is proved')
else :
    print ('Alternate Hypothesis is proved')
```

We conclude that ,  Different zones generate the different amount of sales.
